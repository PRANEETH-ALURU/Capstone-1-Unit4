### Categorical data

```python
# Re-import necessary libraries and dataset due to session reset
import pandas as pd
from scipy.stats import chi2_contingency

# Convert 'Attrition' to numerical for correlation and analysis
df['Attrition_Num'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# Define categorical columns for chi-square analysis
categorical_columns = ['Gender', 'MaritalStatus', 'EducationField', 'BusinessTravel', 'Department', 'OverTime']

# Function to perform chi-square test on all categorical columns vs Attrition
def chi_square_test(data, target_column, categorical_columns):
    chi_square_results = {}

    for column in categorical_columns:
        # Create a contingency table
        contingency_table = pd.crosstab(data[column], data[target_column])

        # Perform chi-square test
        chi2, p, dof, expected = chi2_contingency(contingency_table)

        # Store results
        chi_square_results[column] = {'Chi2': chi2, 'p-value': p, 'Degrees of Freedom': dof}

    # Convert results to DataFrame for easy viewing
    return pd.DataFrame(chi_square_results).transpose()

# Perform chi-square test
chi_square_results_df = chi_square_test(df, 'Attrition', categorical_columns)

# Display the results
#Removed the import for the unavailable 'ace_tools' and used the print function instead
print("Chi-Square Test Results for Categorical Features")
print(chi_square_results_df)
```

The chi-square test results show that **Marital Status**, **Education Field**, **Business Travel**, **Department**, and **OverTime** are significantly associated with employee attrition (p-value < 0.05). This indicates that these categorical variables have a significant relationship with whether an employee leaves the company. Specifically, **OverTime** has the strongest relationship with attrition (p-value close to 0), suggesting that employees working overtime are much more likely to leave. Conversely, **Gender** is not significantly associated with attrition (p-value > 0.05), indicating that it does not play a significant role in employee turnover.

### Through IQR

```python
import pandas as pd

# Load the dataset (replace 'file_path' with your file path)
file_path = "HR-Analytics.csv"
data = pd.read_csv(file_path)

# Select numerical columns
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns

# Display the numerical features
print("Numerical Features:\n", numerical_columns)

# Show the first few rows of the numerical data
numerical_data = data[numerical_columns]
```

```python
def detect_outliers_iqr(df, columns):
    outliers = {}
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    return outliers

# Identify outliers using the IQR method
outliers_iqr = detect_outliers_iqr(data, numerical_columns)

# Plot boxplots for the numerical columns to visualize outliers
plt.figure(figsize=(20, 15))
for i, col in enumerate(numerical_columns):
    plt.subplot(6, 6, i + 1)
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)

plt.tight_layout()
plt.show()

# Display the columns that contain outliers
outliers_summary = {col: len(outliers_iqr[col]) for col in outliers_iqr if len(outliers_iqr[col]) > 0}
print("Number of Outliers in Each Column:\n", outliers_summary)

# Function to cap outliers using IQR method
def cap_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
    return df

# Apply capping on the columns with detected outliers
data_capped = cap_outliers(data.copy(), outliers_summary.keys())

# Confirm that capping has been applied by checking for remaining outliers
remaining_outliers = detect_outliers_iqr(data_capped, outliers_summary.keys())

# Summary of remaining outliers
remaining_outliers_summary = {col: len(remaining_outliers[col]) for col in remaining_outliers if len(remaining_outliers[col]) > 0}
print("Remaining Outliers After Capping:\n", remaining_outliers_summary)
```

We are handling the outliers by analysing them using the box plots and setting the inter-quartile range of the upper and lower half as limits. We are reducing the values to lower and upper bound any point that crossess those limits. After capping them, we can see the outliers count is decreased and no outliers left at the last.
